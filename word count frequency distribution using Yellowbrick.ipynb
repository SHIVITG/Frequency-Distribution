```python
# Visualization : To plot frequency distribution of words of a text file using YELLOWBRICK
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from yellowbrick.text import FreqDistVisualizer
from yellowbrick.datasets import load_hobbies
```

```python
# Raw Data
```

```python
corpus = open('Merge.txt','r+')
vectorizer = CountVectorizer()
docs = vectorizer.fit_transform(corpus.readlines())
features = vectorizer.get_feature_names()
visualizer = FreqDistVisualizer(features=features, orient = 'v', n= 50, size=(1080, 1080))
visualizer.fit(docs)
visualizer.show(outpath="raw.pdf")
```

```python
# Cleaned data without stopwords
```

```python
corpus = open('test_sw.txt','r+')
vectorizer = CountVectorizer()
docs = vectorizer.fit_transform(corpus.readlines())
features = vectorizer.get_feature_names()
visualizer = FreqDistVisualizer(features=features, orient = 'v', n= 50, size=(1080, 1080))
visualizer.fit(docs)
visualizer.show(outpath="cleaned.pdf")
```
